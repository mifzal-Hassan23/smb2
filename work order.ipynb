```python
import pandas as pd
import datatable as dt 
import numpy as np
from datetime import date
import datetime as dt
import re 
import dateutil.relativedelta
from pandas.tseries.offsets import Day, MonthEnd
wo_df=pd.read_excel(r'C:\Users\wajiz.pk\Downloads\Mifzal_files\MDI_Automated_Reports\Work_Order\Work_Order.xlsx',sheet_name='Sheet1')
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
wo_df.dropna(axis=0, how='all',inplace=True)
wo_df.dropna(axis=1, how='all',inplace=True)
wo_df.columns=['00','01','02','03','04','05','06','07', '08','09','010','011','012','013','014','015','016','017','018','019','020','021']
wo_df.insert(0, 'Month', np.nan)
wo_df['Month']=np.where(wo_df['00'].str.contains(pat='Exported On:',regex=True)==True,wo_df['00'],wo_df['Month'] )
wo_df['Month']=wo_df['Month'].str.split(" ",2,expand=True)[2]
wo_df['Month']=wo_df['Month'].str.split(" ",1,expand=True)[0]
wo_df["Month"].fillna(method='ffill', inplace = True)
wo_df['Month']=pd.to_datetime(wo_df['Month'])
wo_df.columns=['Month','Property','Priority','Work Order Type','Home Warranty Expiration','Work Order Number'
               ,'Job Description','Instructions', 'Status','Vendor','Unit','Primary Resident','Created At','Estimate Req On'
               ,'Estimated On','Estimate Amount','Scheduled Start','Scheduled End','Work Completed On','Completed On'
               ,'Amount','Invoice','Recurring']

wo_df = wo_df.iloc[11:]
wo_df.insert(3,'Propert', np.nan)


wo_df['Propert']=np.where((wo_df['Work Order Type'].isnull()) 
                             & (wo_df['Home Warranty Expiration'].isnull()) 
                             & (wo_df['Work Order Number'].isnull()) 
                             & (wo_df['Job Description'].isnull())
                             & (wo_df['Instructions'].isnull())
                             & (wo_df['Status'].isnull())
                             & (wo_df['Vendor'].isnull())
                             & (wo_df['Unit'].isnull())
                             & (wo_df['Primary Resident'].isnull())
                             & (wo_df['Created At'].isnull())
                             & (wo_df['Estimate Req On'].isnull())
                             & (wo_df['Estimated On'].isnull())
                             & (wo_df['Estimate Amount'].isnull())
                             & (wo_df['Scheduled Start'].isnull())
                             & (wo_df['Scheduled End'].isnull())
                             & (wo_df['Work Completed On'].isnull())
                             & (wo_df['Completed On'].isnull())
                             & (wo_df['Amount'].isnull())
                             & (wo_df['Invoice'].isnull())
                             & (wo_df['Recurring'].isnull()),wo_df['Propert'],wo_df['Property']) 
wo_df['Propert']=wo_df['Propert'].str.split('-',-1,expand=True)[0]
wo_df['Propert'] = wo_df['Propert'].str.replace('Apartments',' ')
wo_df["Propert"].fillna(method='ffill', inplace = True)
wo_df.dropna(subset=['Priority'],inplace=True )
wo_df['Property']=wo_df['Propert']
wo_df.rename(columns = {'Propert':'Property2'},inplace=True)
neworder = ['Month','Property','Property2','Priority','Work Order Type','Home Warranty Expiration','Work Order Number'
               ,'Job Description','Instructions', 'Status','Vendor','Unit','Primary Resident','Created At','Estimate Req On'
               ,'Estimated On','Estimate Amount','Scheduled Start','Scheduled End','Work Completed On','Completed On'
               ,'Amount','Invoice','Recurring']
wo_df = wo_df.reindex(columns=neworder)
wo_df['Month']= wo_df['Month'] - pd.DateOffset(months=1)
wo_df['Month']=pd.to_datetime(wo_df['Month'])+MonthEnd(1)
wo_df=wo_df.replace(' ', '')

wo_df


# ////////////////////////////////////////////////////////////////////////////////////////////////////////////////////

# writer = pd.ExcelWriter('Work_order.xlsx', engine='xlsxwriter')
# wo_df.to_excel(writer, sheet_name='workorders', startrow=1, header=False, index=False)
# workbook = writer.book
# worksheet = writer.sheets['workorders']
# (max_row, max_col) = wo_df.shape
# column_settings = [{'header': column} for column in wo_df.columns]
# worksheet.add_table(0, 0, max_row, max_col - 1, {'columns': column_settings,'name': "UVD"})
# worksheet.set_column(0, max_col - 1, 12)
# writer.save()
```
